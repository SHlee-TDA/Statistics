# Chapter 2. Some Probabilistic Concepts and Results

## 2.1 Probability Functions and Some Basic Properties and Results

- *Experiment (실험)* : 특정 조건 하에서 수행되는 절차를 묘사하여 동일한 조건 하에서 절차를 몇 번이고 반복할 수 있으며, 절차가 완료되면 특정 결과가 관찰되는 것.
  - *Deterministic experiment* : 실험이 수행되는 조건 하에서 그 결과가 항상 동일하게 나오는 실험.
  - *Random experiment* : 실험이 수행되는 조건 하에서 그 결과가 결정적이지 않고, 알려진 가능한 결괏값의 집합의 원소 중 하나로 나오는 실험. 앞으로의 experiment는 항상 random experiment만을 이야기하도록 하자.

- *Sample space $S$* : random experiment에서 얻을 수 있는 가능한 모든 결과값의 집합
- *Sample point &s&* : Sample space의 원소
- *Events $E$* : sample space의 부분집합
  - *Simple event $\left\{s\right\}* : 하나의 sample point로 이루어진 event
  - *Composite event* : 두 개 이상의 sample point로 이루어진 event
- Sample space는 $\sigma$-field 구조를 가진다.
- random experiment를 통해서 결과 $s \in A$를 얻을 경우 event $A$가 *occur* 또는 *happen* 했다고 한다.


**Definition 1**
A *probability function (or probability measure)* denoted by $P : \mathcal{A} \rightarrow [0,1]$ is a (set) function which assigns to each event $A$ a number denoted by $P(A)$, called the *probability of $A$*, and satisfies the following requirements:
- (P1) $P$ is *non-negative*; that is, $P(A) \geq 0$, for every event $A$.
- (P2) $P$ is *normed*; that is, $P(S)$ = 1.
- (P3) $P$ is *$\sigma$-additive*; that is, for every collection of pairwise (or mutually) disjoint events $A_j, j=1,2,\ldots$, we have $P(\sum_j A_j) = \sum_j P(A_j)$.

This is the axiomatic (Kolmogorov) definition of probability.
The triple $(S, \mathcal{A}, P)$ is known as a *probability space*.

*Remark 1*
If $S$ is finite, then every subset of $S$ is an event (that is, $\mathcal{A}$ is taken to be the discrete $\sigma$-field).
In such a case, there are only finitely many events and hence, in particular, finitely many pairwise disjoint events.
Then (P3) is reduced to:
- (P3') $P$ is *finitely additive*; that is, for every collection of pairwise disjoint events, $A_j, j=1,2,\ldots, n$, we have $$P\left(\sum_{j=1}^n A_j\right) = \sum_{j=1}^n P(A_j).$$

### 2.1.1 Consequences of Definition 1

- (C1) $P(\empty) = 0$. (Any event, possibly $\neq \empty$, with probability 0 is called a *null* event.)
- (C2) $P$ is *finitely additive; that is for any event $A_j , j=1,2,\ldots,n$ such that $A_i \cap A_j = \empty, i\neq j$, $P(\sum_{j=1}^n A_j) = \sum_{j=1}^n P(A_j)$.
- (C3) For every event $A$, $P(A^c) = 1 - P(A)$.
- (C4) $P$ is a *non-decreasing function*; that is $A_1 \subseteq A_2$ implies $P(A_1) \leq P(A_2)$.
  
*Remark 2* If $A_1 \subseteq A_2$, then $P(A_2 - A_1) = P(A_2) - P(A_1)$, but *this is not true, in general*.

- (C5) $0 \leq P(A) \leq 1$ for every event $A$.
- (C6) For any events $A_1, A_2$, $P(A_1 \cup A_2) = P(A_1) + P(A_2) - P(A_1 \cap A_2)$.
- (C7) $P$ is *subadditive*; that is $P(\cup_{j=1}^\infty A_j) \leq \sum_{j=1}^\infty P(A_j)$ and also $P(\cup_{j=1}^n A_j) \leq \sum_{j=1}^n P(A_j)$.


Let $S = \left\{s_1,s_2, \ldots, s_n \right\}$, let $\mathcal{A}$ be the discrete $\sigma$-field of $S$, and define $P$ as $P(\left\{s_j\right\}) = 1/n, j=1,2,\ldots,n$.
With this definition, $P$ clearly satisfies (P1) - (P3') and this is the *classical* definition of probability.
Such a probability function is called a *uniform probability function*.
This definition is adequate as long as $S$ is finite and the simple events $\left\{s_j\right\}, j=1,2,\ldots,n$, may be assumed to be "equally likely", but it breaks down if either of these two conditions is not satisfied.
However, this classical definition together with the following *relative frequency (or statistical)* definition of probability served as a motivation for arriving at the axioms (P1) - (P3) in the Kolmogorov definition of probability.

The relative frequency definition of probability is this :
Let $S$ be any sample space, finite or not, supplied with a class of events $\mathcal{A}$.
A random experiment associated with the same sample space $S$ is carried out $n$ times.
Let $n(A)$ be the number of times that the event $A$ occurs.
If, as $n\to \infty$, $\lim[n(A)/n]$ exists, it is called the probability of $A$, and is denoted by $P(A)$.
Clearly, this definition satisfies (P1), (P2) and (P3').


**Theorem 1 (Additive Theorem)**
For any finite number of events, we have 
$$P\left(\cup_{j=1}^n A_j\right) = \sum_{j=1}^n P(A_j) - \sum_{1\leq j_1 < j_2 n} P(A_{j_1} \cap A_{j_2}) + \sum_{1\leq j_1 < j_2 < j_3 \leq n} P(A_{j_1}\cap A_{j_2}\cap A_{j_3})- \cdots + (-1)^{n+1} P(A_1 \cap A_2 \cap \cdots A_n).$$


**Theorem 2** 
Let $\left\{A_n\right\}$ be a sequence of events such that, as $n\to \infty$, $A_n \uparrow$ or $A_n \downarrow$.
Then, $$P\left(\lim_{n\to\infty}A_n\right) = \lim_{n\to \infty} P(A_n).$$

## 2.2 Conditional Probability

**Definition 2**
Let $A$ be an event such that $P(A) > 0$.
Then the conditional probability, given $A$, is the (set) function denoted by $P(\cdot | A)$ and defined for every event $B$ as follows:
$$ P(B|A) = \frac{P(A\cap B)}{P(A)}.$$
$P(B|A)$ is called the *conditional probability* of $B$, given $A$.

**Theorem 3 (Multiplicative Theorem)**
Let $A_j, j=1,2,\ldots,n$, be events such that $$P\left(\cap_{j=1}^{n-1} A_j\right)>0.$$ 
Then $$ P\left(\cap_{j=1}^{n} A_j\right) = P(A_n | A_1 \cap A_2 \cap \cdots \cap A_{n-1}) \times P(A_{n-1} | A_1 \cap \cdots A_{n-2}) \cdots P(A_2 |A_1)P(A_1).$$

*Remark 3*
The value of the above formula lies in the fact that, in general, it is easier to calculate the conditional probabilities on the right-hand side.

Let $A_j, j=1,2,\ldots$, be events such that $A_i \cap A_j = \empty, i\neq j$, and $\sum_j A_j = S$.
Such a collection of events is called a *partition* of $S$.
The partition is *finite* or (denumerably) *infinite*, accordingly, as the events $A_j$ are finitely or (denumerably) infinitely many.
For any event, we clearly have: $$ B = \sum_j (B\cap A_j).$$
Hence $$P(B) = \sum_j P(B\cap A_j) = \sum_j P(B|A_j)P(A_j),$$ provided $P(A_j) > 0$, for all $j$.
Thus we have the following theorem.

**Theorem 4 (Total Probability Theorem)**
Let $\left\{A_j, j=1,2,\ldots\right\}$ be a partition of $S$ with $P(A_j) > 0$, for all $j$.
Then for $B \in \mathcal{A}$, we have $P(B) = \sum_j P(B|A_j)P(A_j)$.

This formula gives a way of evaluating $P(B)$ in terms of $P(B|A_j)$ and $P(A_j), j=1,2,\ldots$.
Under the condition that $P(B)>0$, the above formula can be "reversed" to provide an expression for $P(A_j|B), j=1,2,\ldots$. 
In fact, $$P(A_j|B) = \frac{P(A_j\cap B)}{P(B)} = \frac{P(B|A_j)P(A_j)}{P(B)} = \frac{P(B|A_j)P(A_j)}{\sum_i P(B|A_i)P(A_i)}.$$

**Theorem 5 (Bayes Formula)**
If $\left\{A_j, j=1,2,\ldots\right\}$ is a partition of $S$ and $P(A_j) > 0, j=1,2,\ldots,$ and if $P(B)>0$, then $$P(A_j|B) = \frac{P(B|A_j)P(A_j)}{\sum_i P(B|A_i)P(A_i)}.$$

*Remark 4* 
It is important that one checks to be sure that the collection $\left\{A_j,j\geq1 \right\}$ forms a partition of $S$, as only then are the above theorems true.

**EXAMPLE 2**
5지선다 문제로 구성된 숙제가 있다고 하자.
한 학생이 숙제를 풀었다면, 학생은 정답을 알고 있을 것이다.
그렇지 않으면 그 정답을 찍은 것이다.
$p$를 학생이 숙제를 풀어서 낸 사건 $A$의 확률이라고 하고, $B$는 학생이 정답을 정확히 맞춘 사건이라고 하자.
이때 조건부 확률 $P(A|B)$를 $p$를 이용해 표현하라.

*Solution)*
$A$와 $A^c$가 sample space $S$의 partition을 형성함에 주목하면, Theorem 4와 Theorem 5에 의해,
$$P(A|B) = \frac{P(B|A)P(A)}{P(B|A)P(A) + P(B|A^c)P(A^c)} = \frac{1\cdot p}{1\cdot p + \frac{1}{5}\cdot(1-p)} = \frac{5p}{4p+1}.$$


Consider two partitions $\left\{A_i, i=1,2,\ldots\right\}, \left\{B_j,j=1,2,\ldots\right\}$.
Then, clearly, 
$$ A_i = \sum_j (A_i \cap B_j), i=1,2,\ldots,$$
$$ B_j = \sum_i (B_j \cap A_i), j=1,2,\ldots,$$
and
$$\left\{A_i\cap B_j, i=1,2,\ldots, j=1,2,\ldots\right\}$$
is a partition of $S$.
In fact, 
$$(A_i \cap B_j) \cap (A_{i'}\cap B_{j'}) =\empty \text{ if } (i,j) \neq (i',j')$$
and
$$\sum_{i,j} (A_i \cap B_j) = \sum_i\sum_j (A_i\cap B_j) = \sum_i A_i = S.$$
The expression $P(A_i \cap B_j)$ is called the *joiint probability* of $A_i$ and $B_j$.
On the other hand, from
$$ A_i = \sum_j (A_i \cap B_j) \text{ and } B_j = \sum_i (B_j \cap A_i),$$
we get
$$P(A_i) = \sum_j P(A_i \cap B_j) = \sum_j P(A_i|B_j)P(B_j),$$
provided $P(B_j)>0, j=1,2,\ldots$, and
$$P(B_j) = \sum_i P(B_j \cap A_i) = \sum_i P(B_j|A_i)P(A_i),$$
provided $P(A_i)>0, i=1,2,\ldots$.
The probabilities $P(A_i), P(B_j)$ are called *marginal probabilities*.
We have analogous expressions for the case of more than two partitions of $S$.
